```python
import time
import numpy as np
import pandas as pd
import matplotlib as mat
from matplotlib import pyplot as plt
from sklearn.datasets import load_iris
```

```python
# importing the dataset
iris = load_iris()
df = pd.DataFrame(iris.data, columns = iris.feature_names)
print(len(df))
df.head()
```

# Extraction and data transformation

```python
# including the target values inside the dataframe
df['target'] = iris.target
df.head()
```

```python
# including the target names values inside the dataframe
df['species'] = pd.Categorical.from_codes(iris.target, iris.target_names)
df.head()
```

```python
# separete the data by species
setosa = df[df['species']== setosa]
versicolor = df[df['species']== versicolor]
virginica = df[df['species']== virginica]
```

```python
# calculate the mean for each feature
df.groupby('species').mean().T
```

```python
# scatter Plot of sepal length versus line number 
plt.figure(figsize=(12, 8), dpi=80)
plt.scatter(range(len(df)), df['petal width (cm)'], c=df['target'])
plt.xlabel('line number', fontsize=15)
plt.ylabel('Sepal length (cm)', fontsize=15)
plt.title('Dispersion of atributes, colored by marks of class target', fontsize=15)
plt.show()
```
