```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import chisquare
```

```python
#CSPNe com atmosfera irradiando como corpo negro
df = pd.read_csv('Aller_vs_Peimbert_01_04.csv', delimiter =' ')
#df = df.drop(11,axis = 0) #optei por retirar essa linha para que os pontos fiquem mais pr√≥ximos
df
```

```python
x = df.iloc[:,3].astype(float)
y = df.iloc[:,1].astype(float)
```

```python
X =np.linspace(0,400)
Y =np.linspace(0,400)
a = pow(10,-3)
X1 =np.linspace(a,200)
Y1 =np.linspace(a,300)
```

```python
fig, (ax1) = plt.subplots(1, 1,figsize=(10,10),constrained_layout=True)
fig.suptitle('Rela√ß√£o (1x1) entre as medidas de $I({\lambda})$/$I(H_{\u03B2})$', fontsize=18, color='black')
#gr√°fico 1
ax1.scatter(x,y,marker=(5, 1), color= 'blue' )
ax1.plot(X,Y, color= 'black', linestyle = '--',linewidth = 0.8 )

ax1.set(xlabel="PEIMBERT el al. (1976)", ylabel= "ALLLER et al. (1995)")

#for ax in fig.get_axes():
#    x.label_outer()
#plt.savefig('Relacao_1x1_bb47kK_powerlaw.png')
plt.show()
```
